# Time series

A time series is any data set created with values grouped by different points in time. In this tutorial, we will learn about some of the time series tools that the pandas library has to offer. We will also plot some nice charts and learn how to format them. 

Time series can be spaced with different frequencies:  
   - Hourly weather measurements  
   - Daily counts of covid cases
   - Monthly energy consumption
   - Irregular timestamped data of 911 emergency calls  

Pandas library allow us to handle either type of time series.

## The data set: Victorian COVID cases by age group and Local Government Area (LGA)  

We will be working with the data sets provided by the Victorian Department of Health. In particular, two sets:

1. [All Victorian PCR SARS-CoV-2 cases by age group](https://discover.data.vic.gov.au/dataset/all-victorian-sars-cov-2-cases-by-age-group)
2. [All Victorian PCR SARS-CoV-2 cases by LGA](https://discover.data.vic.gov.au/dataset/all-victorian-sars-cov-2-cases-by-local-government-area)

### Data wrangling and munging    

**Data Wrangling** or **Data Munging** is the preparation of your raw data for a specific purpose. This means transforming the data into the format better-suited for manipulation.  

To start with the data munging we should know how the information has been stored. In our case, Covid cases are reported in a data set where each row contains a single PCR SARS-CoV-2 case for either the given age group (bands of ten years) or the lga.  

Let's take a look at the data sets...  

```python
import pandas as pd

# Reading the files
covid_age = pd.read_csv('NCOV_COVID_Cases_by_Age_Group_20220916_0820.csv')
covid_lga = pd.read_csv('NCOV_COVID_Cases_by_LGA_20220916_0820.csv')

print(covid_age.head(10))
print(covid_lga.head(10))
```

```python
# Size of the data sets
print('Size of covid_age = ', covid_age.shape)
print('Size of covid_lga = ',covid_lga.shape)
```

Now we know that there is a total of 1,148,144 cases in our data sets. However, the format used does not allow us to know directly how many cases were reported for the same group on the same date.  

Let's organise our dataframe so it groups and shows the number of cases by date and age group!

```python
# First we have to create an extra column to store the number of cases

# We will pre-fill it with ones since every row is a single case
covid_age['n_cases'] = 1
covid_lga['n_cases'] = 1

# Then, we have to group the data by date and age group 
# and count the number of times we have a specific group in the dataframe
# for this, we use groupby() and count() methods
covid_age = covid_age.groupby(['diagnosis_date','agegroup']).n_cases.count().reset_index()
covid_lga = covid_lga.groupby(['diagnosis_date','Localgovernmentarea']).n_cases.count().reset_index()

# Beacuse we grouped our data, the number of rows has decreased and that is why we reset our index
# Let's take a look at our new dataframes
print(covid_lga)
print(covid_age)
```

```python
# Let's check our data types 
print(covid_lga.dtypes)
print(covid_age.dtypes)
covid_age['diagnosis_date']
```

The main pandas data structures for working with dates and times is a **Timestamp**. We can use the to_datetime() function to create Timestamps from a wide variety of date/time formats to a pandas datetime object.

```python
covid_age['diagnosis_date'] = pd.to_datetime(covid_age['diagnosis_date'])
covid_lga['diagnosis_date'] = pd.to_datetime(covid_lga['diagnosis_date'])

# to check what is the sequence used for indexing
print(covid_age.dtypes)
print(covid_lga.dtypes)
```

However, in our dataframes, not all the age groups neither the lga reported cases on a daily base. Also, not all the groups or lga reported cases on the same dates and therefore, we have irregular timestamps.  

To manipulate our data and make it comparable within the same time series, we must create new timestamps with zero cases for those groups where no cases were reported on an existent date in the dataframe. 

```python
# Creating a dataframe organised by date and agegroup as indexes of the tabular data

covid_age = (covid_age.set_index(['diagnosis_date','agegroup'])['n_cases']
             .unstack(fill_value=0))

covid_age
```

```python
# Creating a dataframe organised by date and lga as indexes of the tabular data

covid_lga = (covid_lga.set_index(['diagnosis_date','Localgovernmentarea'])['n_cases']
             .unstack(fill_value=0))

# (S) = Shire Council, (RC) = Rural City Council
# (C) = City Council, etc...
covid_lga
```

Now we have our data organised by the number of cases for the overall reported dates!  

What if we want to retrieve the information for a given date?

```python
# Using the loc() method will give us the number of cases on a determined date
print(covid_age.loc['2022-09-11'])
print(covid_lga.loc['2022-09-11'])
```

Now, let's visualize what we have!! Starting with the age group dataframe...

```python
import matplotlib.pyplot as plt
import seaborn as sns              # Using seaborn styling for our plots

print(covid_age['20-29'])
sns.set(rc={'figure.figsize':(11, 6)})  
sns.set(style="whitegrid", font_scale = 1)

covid_age.plot(linewidth=0.5);
plt.legend(fontsize=12, title='Age group', title_fontsize=12)
```

From what we see in above figure, a logarithmic plot would be more appropriate...

```python
# Plotting our data using a logarithmic scale for y axis
covid_age.plot(linewidth=0.5);
plt.gca().set_yscale('log')
```

But maybe it is a bit too much to plot all the data together!! However, from above figure we can spot that the age group with the highest number of cases over time is the 20-29 age group!  

Let's compare this group to those around it.

```python
# Age groups to compare
cols_plot = ['10-19_', '20-29', '30-39', '40-49']

# Getting the maximum number of cases over the complete time series
# (we will use it to fix our y-axis limit)
max_ncases = covid_age.max()
print(max_ncases)
max_allgroups = max_ncases.max()
print('max number of cases of all groups = ', max_allgroups)

# Plotting the selected age groups to compare
axes = covid_age[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 10), subplots=True)
for ax in axes:
    ax.set_ylabel('Number of cases')
    ax.set_ylim([None, max_allgroups])
    
```

```python
# But again we need logarithmic scale!

import numpy as np

ylim = np.around(np.log10(max_allgroups),1)
axes = covid_age[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 10), subplots=True)
for ax in axes:
    ax.set_ylabel('Number of cases')
    ax.set_yscale('log')
    ax.set_ylim([None, 10**ylim])
    
```

An useful aspect of the DatetimeIndex is that the individual date/time components are all available as attributes such as year, month, day, and so on.  

We can add them as columns to covid_age to filter the data according to the year, month, and weekday name.

```python
covid_age = covid_age.reset_index()
covid_lga = covid_lga.reset_index()

# Set date as index to add extra columns
covid_age = covid_age.set_index('diagnosis_date')
covid_lga = covid_lga.set_index('diagnosis_date')

# Add columns with year, month, and weekday name
covid_age['Year'] = covid_age.index.year
covid_age['Month'] = covid_age.index.month
covid_age['Weekday Name'] = covid_age.index.day_name()

covid_lga['Year'] = covid_lga.index.year
covid_lga['Month'] = covid_lga.index.month
covid_lga['Weekday Name'] = covid_lga.index.day_name()

covid_age.index.year
```

```python
ax = covid_age.loc['2020','20-29'].plot()
ax.legend(fontsize=12)
ax.set_ylabel('Number of cases')
```

Let's zoom in and look at the peak from July 2020 to September 2020

```python
ax = covid_age.loc['2020-07':'2020-09', '20-29'].plot(marker='o', linestyle='-')
ax.set_ylabel('Number of cases')
```

We can improve the visualisation to show the weekly trend. For this purpose, we can add vertical gridlines on a weekly time scale (instead of on the first day of each month) with matplotlib.dates.

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots()
ax.plot(covid_age.loc['2020-07':'2020-09', '20-29'], marker='o', linestyle='-')
ax.set_ylabel('Number of cases')
ax.set_title('July-Sept 2020 Covid cases 20-29 age grouup')
# Set x-axis major ticks to weekly interval, on Mondays
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))
# Format x-tick labels as 3-letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
```
